**4. Data cleaning**

```python
print(nycdata.isnull().sum())

nycdata.dropna(inplace=True)

print(nycdata.isnull().sum())
```

```python
nycdata.shape
```

```python
nycdata.duplicated().sum()

# print(nycdata[nycdata.duplicated()])

nycdata.drop_duplicates(inplace=True)

print(f'Number of duplicates: {nycdata.duplicated().sum()}')
```

```python
nycdata.dtypes
```

```python
print(nycdata[nycdata['baths'] == 'Not specified'])

print(nycdata['baths'].unique())

nycdata = nycdata.drop(nycdata[nycdata['baths'] == 'Not specified'].index)
print(nycdata['baths'].unique())
nycdata['baths'] = nycdata['baths'].astype(float)
```

```python
nycdata['id'] = nycdata['id'].astype(object)
nycdata['host_id'] = nycdata['host_id'].astype(object)
nycdata.dtypes

```

#### 5. EDA

**Univariate Analysis**

```python
# outliers in price

# Two places available at 100,000. Seem like they are trying to block listings, will be removed
(nycdata[nycdata['price'] > 20000])

df = nycdata[nycdata['price'] < 20000]
# sns.boxplot(data=nycdata, x='price')
bxplt = (sns.boxplot(data=df, x='price'))

# Box plot has very long tail, apply log to get better visualization
bxplt.set_xscale('log')

plt.title('Histogram of price (Log)')
plt.xlabel('Log(Price)')
```

```python
# Look at price distributions
# sns.histplot(data = nycdata, x='price')
nyc_hist = sns.histplot(data = (df), x='price')

# histogram has very long tail, apply log to get better visualization
nyc_hist.set_xscale('log')

plt.title('Histogram of price (Log)')
plt.ylabel('Frequency')
plt.xlabel('log(Price)')
```

```python
nyc_hist = sns.histplot(data = (df), x='availability_365')

plt.title('Histogram of availability')
plt.ylabel('Frequency')

```

```python
# information about price based on neighbourhood_group
df.groupby(by='neighbourhood_group').agg(
    mean_price = ('price', 'mean'),
    median_price = ('price', 'median'),
    min_price = ('price', 'min'),
    max_price = ('price', 'max'),
    num_listings = ('price', 'count')
)
```

**Feature Engineering**

```python
# Price per bed column
df['price_per_bed'] = df['price'] / df['beds']
df.head()
```

```python
# information about price_per_bed based on neighbourhood_group
df.groupby(by='neighbourhood_group').agg(
    mean_price = ('price_per_bed', 'mean'),
    median_price = ('price_per_bed', 'median'),
    min_price = ('price_per_bed', 'min'),
    max_price = ('price_per_bed', 'max'),
    num_listings = ('price_per_bed', 'count')
)
```
